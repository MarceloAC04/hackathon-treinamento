```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

df = pd.read_excel("Despesas_Tratadas.xlsx")
```

```python
df['data'] = pd.to_datetime(df['data'], dayfirst=True)

# Criar uma nova coluna com o mês
df['mes'] = df['data'].dt.to_period('m')

# Agrupar despesas por mês e setor
gastos_mensais_por_setor = df.groupby(['mes', 'setor'])['valor'].sum().unstack().fillna(0)

# Plotar gráfico
plt.figure(figsize=(25, 15))
gastos_mensais_por_setor.plot(kind='bar', stacked=True)
plt.title('Distribuição das Despesas Mensais por Setor')
plt.xlabel('Mês')
plt.ylabel('Valor (R$)')
plt.xticks(rotation=45)
plt.legend(title='Setor', bbox_to_anchor = [1, 0.9], loc='upper left')
plt.tight_layout()
plt.grid(True) 
plt.show()
```
